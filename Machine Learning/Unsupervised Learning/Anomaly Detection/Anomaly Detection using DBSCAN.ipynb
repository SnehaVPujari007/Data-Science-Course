```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
X, y = make_circles(n_samples= 250 ,factor = 0.3 ,noise = 0.1 )
```

```python
plt.scatter(X[:,0] , X[:,1] )
```

```python
dbcan =DBSCAN(eps = 0.10)
```

```python
dbcan.fit_predict(X)
```

```python
dbcan_labels =dbcan.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=dbcan_labels)
```
