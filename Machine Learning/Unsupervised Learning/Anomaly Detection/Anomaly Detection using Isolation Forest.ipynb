```python
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
X, y = make_circles(n_samples= 250 ,factor = 0.3 ,noise = 0.1 )
```

```python
plt.scatter(X[:,0] , X[:,1] )
```

```python
Isolated = IsolationForest(contamination= 0.1)
```

```python
Isolated.fit(X)
```
