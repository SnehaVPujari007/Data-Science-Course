####   Support  Vector  Machine

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
## Creating a dataset

from sklearn.datasets import make_classification
```

```python
make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,
                          n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0],y = pd.DataFrame(X)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test , y_train,y_test = train_test_split(X,y,test_size = 0.25 , random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred1 = rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomail = SVC(kernel='poly')
polynomail .fit(X_train,y_train)
y_pred2 = polynomail .predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid1 = SVC(kernel='sigmoid')
sigmoid1.fit(X_train,y_train)
y_pred3 =  sigmoid1.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid correctly
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],  
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
}
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Create an SVM model
svm_model = SVC()

# Perform GridSearchCV
grid = GridSearchCV(estimator=svm_model, param_grid=param_grid, refit = True , cv = 5 ,verbose = 3)

# Fit the model
grid.fit(X_train, y_train)

y_pred34=  grid.predict(X_test)
print(classification_report(y_test,y_pred34))
print(confusion_matrix(y_test,y_pred34))



# Display best parameters and score
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```
