Word Emebedding (Feature Extraction)

```python
from tensorflow.keras.preprocessing.text import one_hot
import numpy as np

```

```python
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
##  Define the vocabulary size
voc_size = 10000
```

```python
### One Hot Representation
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
##  word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences # ("as embedded layer")
from tensorflow.keras.models import Sequential
```

```python
# Making every sentence of equal words
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)   # padding = "post"
print(embedded_docs)

```

```python
## feature representation
dim = 10

```

```python
model = Sequential()
model.add(Embedding(voc_size, dim , input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.build(input_shape=(None,sent_length))
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

```python
model.predict(embedded_docs[[0]]) # Pass a 2D array by using slicing
```
