```python
import numpy as np

## create a array using numpy 
## create a 1D array

arr1 = np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
array1 = np.array([[1,2,3],[4,5,6]])

print("Array: \n",array1)
print("Shape: " , array1.shape)
print("Number of dimensions :" , array1.ndim)
print("Number of size :" , array1.size)
print("Data Type:" ,array1.dtype)
print("Item size :" , array1.itemsize)

```

```python
###  Numpy Vectorized Operation

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,20,30,40,50])

### Element Wise addition

print("Addition:" , arr1+arr2)


###  Element Wise substraction


print("Addition:" , arr1-arr2)


### Element Wise multiplication

print("Multiplication:" , arr1*arr2)


##3 Element wise divison

print("Division :" ,arr1/arr2)
```

```python
## universal Function 

arr = np.array([2,3,4,5,6])

## Square root 

print(np.sqrt(arr))

## Exponential

print(np.exp(arr))


## Sine

print(np.sin(arr))

## natural log 

print(np.log)

```

```python
##  Merging and joining DataFrame 

df1 = pd.DataFrame({'Key' :['A','B','C'] , 'Value1' : [1,2,3]})
df2 = pd.DataFrame({'Key' : ['A','B','D'],'Value2' : [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
##  Merge DataFrame on the 'Key Columns'
pd.merge(df1,df2,on="Key" ,how = "inner")
```

```python
pd.merge(df1,df2,on="Key" ,how = "outer")
```

```python
pd.merge(df1,df2,on="Key",how = "left")
```

```python
pd.merge(df1,df2,on="Key" ,how = "right")
```
