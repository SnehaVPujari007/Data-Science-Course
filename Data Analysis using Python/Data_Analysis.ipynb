```python
array1 = np.array([[1,2,3],[4,5,6]])

print("Array: \n",array1)
print("Shape: " , array1.shape)
print("Number of dimensions :" , array1.ndim)
print("Number of size :" , array1.size)
print("Data Type:" ,array1.dtype)
print("Item size :" , array1.itemsize)

```

```python
###  Numpy Vectorized Operation

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,20,30,40,50])

### Element Wise addition

print("Addition:" , arr1+arr2)


###  Element Wise substraction


print("Addition:" , arr1-arr2)


### Element Wise multiplication

print("Multiplication:" , arr1*arr2)


##3 Element wise divison

print("Division :" ,arr1/arr2)
```

```python
## universal Function 

arr = np.array([2,3,4,5,6])

## Square root 

print(np.sqrt(arr))

## Exponential

print(np.exp(arr))


## Sine

print(np.sin(arr))

## natural log 

print(np.log)

```

```python
x=[1,2,3,4,5]
y = [1,4,9,16,25]


## Create a line plot 
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')

plt.title("Basic Line plot ")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

## Create a customize line plot 


plt.plot(x,y,color='red',linestyle='--',marker = 'o',linewidth = 3 , markersize = 9)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.grid(True)
```

```python
## Multiple Plots 
## Sample Data 

x = [1,2,3,4,5]
y1 = [1,4,8,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize = (9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color = 'green')
plt.title("PLot 1")


plt.figure(figsize = (9,5))
plt.subplot(2,2,2)
plt.plot(y1,x,color = 'blue')
plt.title("PLot 2")


plt.figure(figsize = (9,5))
plt.subplot(2,2,3)
plt.plot(x,y2,color = 'red')
plt.title("PLot 3")

plt.figure(figsize = (9,5))
plt.subplot(2,2,4)
plt.plot(x,y2,color = 'yellow')
plt.title("PLot 4")
```

##  Data Visualization with Seaborn

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
### Basic Plotting with Seaborn
tips =sns.load_dataset('tips')
tips
```

```python
## Create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()

```

```python
##Line plot

sns.lineplot(x='size',y='total_bill', data = tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
## Violin Plot

sns.violinplot(x='day', y = 'total_bill',data = tips)
```

```python
### Histograms 

sns.histplot(tips['total_bill'],bins=10,kde = True)
```

```python
## KDE Plot 
sns.kdeplot(tips['total_bill'], fill = True)
```

```python
## Pairplot
sns.pairplot(tips)
```

```python
## HEatmap
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap = 'coolwarm')
```
